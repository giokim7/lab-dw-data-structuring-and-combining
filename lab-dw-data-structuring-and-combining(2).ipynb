```python
# Your code goes here
import pandas as pd
#save dataframe
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

df1.head()
```

```python


def rename_columns(df):
    df = df.rename(columns={'ST': 'State', 'GENDER': 'Gender'})
    return df

def clean_invalid_values(df):
    gender_mapping = {
        'F': 'Female',
        'M': 'Male',
        'Femal': 'Female',
        'Male': 'Male',
        'female': 'Female'
    }
    df['Gender'] = df['Gender'].replace(gender_mapping)
    return df

def format_dataset(df):
    df["Customer"] = df["Customer"].astype(str)
    df["State"] = df["State"].astype(str)
    df["Gender"] = df["Gender"].astype(str)
    df["Education"] = df["Education"].astype(str)
    df["Income"] = df["Income"].astype(float)
    df["Monthly Premium Auto"] = df["Monthly Premium Auto"].astype(float)
    df["Policy Type"] = df["Policy Type"].astype(str)
    df["Vehicle Class"] = df["Vehicle Class"].astype(str)
    df["Total Claim Amount"] = df["Total Claim Amount"].astype(float)
    return df

def drop_nulls(df):
    df = df.dropna()
    return df

def dealing_duplicates(df):
    df = df.drop_duplicates()
    return df

def main(df):
    # Rename columns   
    df = rename_columns(df)

    # Clean invalid values
    df = clean_invalid_values(df)

    # Format the dataset
    df = format_dataset(df)

    # Drop null values
    df = drop_nulls(df)

    # Deal with duplicates
    df = dealing_duplicates(df)

    # Return the processed DataFrame
    return df

# Call the main function for each DataFrame
df1 = main(df1)
df2 = main(df2)
df3 = main(df3)

# Combine the datasets
combined_data = pd.concat([df1, df2, df3])

# Print the combined dataset
print(combined_data)


```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

3. You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table. 

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
#import file
df = pd.read_csv('marketing_customer_analysis.csv')
import pandas as pd

# Import the file
df = pd.read_csv('marketing_customer_analysis.csv')

# Exercise 2.1: Total revenue by sales channel
total_revenue_pivot = df.pivot_table(index='Sales Channel', values='Total Claim Amount', aggfunc='sum').round(2)
print("Total Revenue by Sales Channel:")
print(total_revenue_pivot)

# Exercise 2.2: Average customer lifetime value by gender and education level
avg_clv_pivot = df.pivot_table(index='Gender', columns='Education', values='Customer Lifetime Value', aggfunc='mean').round(2)
print("\nAverage Customer Lifetime Value by Gender and Education Level:")
print(avg_clv_pivot)

# Exercise 2.3: Number of complaints by policy type and month (long format)
complaints_long = df.pivot_table(index='Policy Type', columns='Month', values='Number of Open Complaints', aggfunc='sum').reset_index().melt(id_vars='Policy Type', var_name='Month', value_name='Number of Complaints')
print("\nNumber of Complaints by Policy Type and Month (Long Format):")
print(complaints_long)

```
